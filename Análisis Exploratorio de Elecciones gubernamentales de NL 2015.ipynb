```python
#Definimos las columnas de interés (partidos políticos)
partidos = ['PAN', 'PRI', 'PRD', 'PT', 'PVEM', 'MC', 'NVA_ALIANZA',
       'PD', 'CC', 'MORENA', 'PH', 'ES', 'CAND_IND1',
       'PRI_PVEM_NVA_ALIANZA_PD', 'PRI_PVEM_NVA_ALIANZA', 'PRI_PVEM_PD',
       'PRI_NVA_ALIANZA_PD', 'PVEM_NVA_ALIANZA_PD', 'PRI_PVEM',
       'PRI_NVA_ALIANZA', 'PRI_PD', 'PVEM_NVA_ALIANZA', 'PVEM_PD',
       'NVA_ALIANZA_PD']
```

```python
#Cantidad de votos por partido
resultadosTotales= resultados2015[partidos].sum().sort_values(ascending=False)
```

```python
#Generamos un pie chart
pieChart=plt.pie(resultadosTotales,labels=resultadosTotales.keys())
```

```python
#Para visualizarlo mejor, podemos tomar únicamente los 5 partidos que mayor cantidad de votos tuvieron
pieChart=plt.pie(resultadosTotales.head(),labels=resultadosTotales.head().keys())
```

```python
#Número de votos por municipio
resultadosPorMun=resultados2015.groupby(by='MUNICIPIO')[partidos].sum()
```

```python
resultadosPorMun.head()
```

```python
#Número de votos por distrito
resultadosPorDistrito = resultados2015.groupby(by='ID_DISTRITO')[partidos].sum()
resultadosPorDistrito
```

```python
#Municipios por distrito
resultados2015.groupby("ID_DISTRITO")["MUNICIPIO"].unique()
```

```python
#O lo inverso, distritos por municipio
resultados2015.groupby("MUNICIPIO")["ID_DISTRITO"].unique()
```

```python
#Normalizacion por distrito - porcentaje de votos
#La función Apply nos permite aplicar una función a todo el dataframe
#lamda es una manera de definir una función sin nombre (normalmente se utiliza cuando no se va a volver a usar)
porDistritoNorm= resultadosPorDistrito.apply(lambda x: 100 * x/resultadosPorDistrito.sum(axis=1))
porDistritoNorm
```

```python
porDistritoNormTOP=porDistritoNorm[porDistritoNorm.apply(lambda x: x>5)].dropna(axis=1)
porDistritoNormTOP
```

```python
sns.heatmap(porDistritoNormTOP,cmap="Blues")
```

```python
forPlot_porDistritoNormTOP = porDistritoNormTOP.apply(lambda x: (x-porDistritoNormTOP.min(axis=1))/(porDistritoNormTOP.max(axis=1)-porDistritoNormTOP.min(axis=1)))
```

```python
sns.heatmap(forPlot_porDistritoNormTOP,cmap="Blues")
```
